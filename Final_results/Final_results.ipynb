# All models results Comparision

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python


# Specify the directory
directory = '/content/'

# Initialize an empty list to store filenames
total = []
all_pickles = []

# Get the list of filenames in the directory
for filename in os.listdir(directory):
  if "pickle" in filename:
    all_pickles.append(filename)

# import the data from pickle files
for i in sorted(all_pickles):
  with open(i, 'rb') as file:
    data = pickle.load(file)
    total.append(data)
```

```python
# Make a DataFrame

model_names = ['Multinomial_Naive_Bayes','Conv1D_with_token_embeddings','pretrained_token_embed','custom_char_embed_conv1d','hybrid_char_token_embed','tribrid_pos_char_token_embed']
all_model_results = pd.DataFrame(total,index = model_names)
```

```python
all_model_results
```

```python
# Reduce the accuracy to same scale as other metrics
all_model_results["accuracy"] = all_model_results["accuracy"]/100
```

```python
# Plot and compare all of the model results
all_model_results.plot(kind="bar", figsize=(10, 7)).legend(bbox_to_anchor=(1.0, 1.0));
```

```python
# Sort model results by f1-score
all_model_results.sort_values("f1", ascending=False)["f1"].plot(kind="bar", figsize=(10, 7));
```
